```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
import preprocess, evaluate, models
from bayesian_classifier_custom import BayesianClassifier
import matplotlib.pyplot as plt

def run_pipeline(normalization_method, reduction_method, runs=5):
    print(f"\n=== Normalization: {normalization_method.upper()} | Reduction: {reduction_method.upper()} ===")

    df = preprocess.load_and_clean_data("games.csv")

    non_numeric_cols = df.select_dtypes(include=['object']).columns
    df = df.drop(columns=non_numeric_cols)

    X = df.drop(columns=["HOME_TEAM_WINS"])
    y = df["HOME_TEAM_WINS"]

    if normalization_method == "none":
        X_processed = X.values
    else:
        X_processed = preprocess.normalize_data(X, method=normalization_method)

    all_models = models.get_models()
    all_models["Bayesian - Multivariate"] = lambda: BayesianClassifier(mode='multivariate')
    all_models["Bayesian - Naive"] = lambda: BayesianClassifier(mode='naive')
    all_models["Bayesian - Nonparametric"] = lambda: BayesianClassifier(mode='nonparametric')

    model_summary = {}
    model_names = []
    mean_accuracies = []

    for name, model_func in all_models.items():
        print(f"\n--- Model: {name} ---")
        accuracies = []
        cumulative_cm = np.zeros((2, 2))
        for i in range(runs):
            X_train, X_test, y_train, y_test = preprocess.split_data(X_processed, y)

            print("Train distribution:", evaluate.class_distribution(y_train))
            print("Test distribution:", evaluate.class_distribution(y_test))

            if reduction_method == "pca":
                X_train = preprocess.apply_pca(X_train)
                X_test = preprocess.apply_pca(X_test)
            elif reduction_method == "sffs":
                X_train, selector = preprocess.apply_sffs(X_train, y_train)
                X_test = selector.transform(X_test)

            model = model_func()
            model.fit(X_train, y_train)
            acc, cm, report = evaluate.evaluate_model(model, X_test, y_test)
            accuracies.append(acc)
            cumulative_cm += cm
            print(f"Run {i+1}: Accuracy = {acc:.4f}")

        mean_accuracy = np.mean(accuracies)
        model_names.append(name)
        mean_accuracies.append(mean_accuracy)
        model_summary[name] = {
            "mean_accuracy": mean_accuracy,
            "std_dev": np.std(accuracies)
        }

        print(f"\n{name}: Accuracy = {mean_accuracy:.4f} ± {np.std(accuracies):.4f}")
        evaluate.plot_confusion_matrix(cumulative_cm, class_names=["Loss", "Win"],
                                       title=f"{name} - Cumulative Confusion Matrix")

    print("\n=== Summary ===")
    for model, stats in model_summary.items():
        print(f"{model}: Accuracy = {stats['mean_accuracy']:.4f} ± {stats['std_dev']:.4f}")

    plt.figure(figsize=(10, 6))
    plt.bar(model_names, mean_accuracies, color='skyblue')
    plt.ylabel('Mean Accuracy')
    plt.xlabel('Model')
    plt.title(f'Model Comparison ({normalization_method.upper()} + {reduction_method.upper()})')
    plt.xticks(rotation=20)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    for norm in ["none", "minmax", "zscore"]:
        for red in ["none", "pca", "sffs"]:
            run_pipeline(norm, red)
```
